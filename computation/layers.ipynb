```python
import torch
import torch.nn.functional as F
from torch import nn

# layer without parameters
class CenteredLayer(nn.Module):
    def __init__(self) -> None:
        super().__init__()

    def forward(self,x):
        return x - x.mean()
```

```python
class MyLinear(nn.Module):
    def __init__(self, in_units, out_units) -> None:
        super().__init__()
        self.weight = nn.Parameter(torch.randn(in_units,out_units))
        self.bias = nn.Parameter(torch.randn(out_units,))

    def forward(self,X):
        linear = torch.matmul(X, self.weight.data) + self.bias.data
        return F.relu(linear)
```

```python
# practice

class VectorizedBilinear(nn.Module):
    def __init__(self, din, dout) -> None:
        super().__init__()
        self.din = din
        self.dout = dout
        self.weight = nn.Parameter(torch.randn(din*din,dout))

    def forward(self,X):
        interaction = torch.bmm(X.unsqueeze(2),X.unsqueeze(1)).view(-1,self.din*self.din)
        out = torch.matmul(interaction,self.weight)
        return out
```
