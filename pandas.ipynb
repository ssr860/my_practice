# autograd

```python
import torch
 
x = torch.arrange(4.0,requires_grad=True)
y = 2 * torch.dot(x,x)

y.backward()
x.grad

# 为了避免梯度累计，计算下一个函数梯度前需要手动清除上一次的值
x.grad.zeros_()
y = x.sum()
y.backward()
x.grad
```

```python
# 若需将某些计算移动到计算图之外，则通过指定新变量阻断梯度回溯
x.grad.zeros_()
y = x*x
u = y.detach()
z = u*x

z.sum().backward()
# 排除了y计算中x的影响

# 仍有单独的导数
x.grad.zero_()
y.sum().backward()
```

```python
# 控制流也可计算梯度如
def f(a):
    b = a * 2
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b
    else:
        c = 100 * b
    return c

a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()
```
